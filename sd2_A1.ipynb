```python
# a) Create a class named Book with the attributes title, author, and copies, and implement the methods borrow_book(), return_book(), and str().

class Book:
  def __init__(self,title,author,copies):
    self.title = title
    self.author = author
    self.copies = copies

# c) A user must not be able to borrow a book if there are no copies left. The program must print a clear message in that case.

  def borrow_book(self):
    if self.copies > 0:
      self.copies -= 1
      print(f"you borrowed {self.title}")
    else:
      print(f"The book {self.title} is unavailable")

  def return_book(self):
    self.copies += 1
    print(f"1 copy of {self.title} was returned")



  def __str__(self):
    return f"The book {self.title} whose author is {self.author} has {self.copies} copies"

book1 = Book("xsdcvfghj","sdfgh",32)
book2 = Book("werty","tfyui",62)

print(book1)
print(book2)

book1.borrow_book()
book1.borrow_book()
book1.borrow_book()

print(book1)

book2.borrow_book()
book2.borrow_book()
book2.borrow_book()
book2.borrow_book()
print(book2)

```

```python
class Book:
    def __init__(self, title, copies=1):
        self.title = title
        self.copies = copies

    def __repr__(self):
        return f"{self.title} ({self.copies} copies)"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title):
        for book in self.books:
            if book.title == title:
                book.copies += 1
                print(f'Updated: "{title}" now has {book.copies} copies')
                return

        new_book = Book(title)
        self.books.append(new_book)
        print(f'Added new book: "{title}"')

# d) If a user searches for a book that does not exist, the program must handle this safely and display a meaningful message.

    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                print(f'Found: {book}')
                return book
        print("Book not found")
        return None


    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                if book.copies > 0:
                    book.copies -= 1
                    print(f'You borrowed "{title}"')
                else:
                    print("No copies available")
                return

        print("Book not found")

    def return_book(self, title):
        for book in self.books:
            if book.title == title:
                book.copies += 1
                print(f'You returned "{title}"')
                return

        print("Book not found")

    def show_all_books(self):
        if not self.books:
            print("Library is empty")
            return

        print("Library books:")
        for book in self.books:
            print(f"- {book}")

# g) Save the final library report into a text file called library_report.txt.

    def save_report(self, filename="library_report.txt"):
        with open(filename, "w") as file:
            file.write("LIBRARY REPORT\n")
            file.write("=================\n\n")

            for book in self.books:
                file.write(f'Title: {book.title}, Copies: {book.copies}\n')

            file.write("\nEnd of report\n")


# e) Create at least 6 Book objects and add them to the library.

lib = Library()
lib.add_book("fghh")
lib.add_book("sdfgh")
lib.add_book("etuu")
lib.add_book("harry potter")
lib.add_book("lord of rings")
lib.add_book("python basics")

lib.find_book("sdfgh")
lib.borrow_book("etuu")
lib.return_book("fghh")

# f) Simulate a complete scenario that includes displaying all books, borrowing several books, trying to borrow the same book until it becomes unavailable, returning a book, and displaying the final library state.
# initial state
lib.show_all_books()

# borrowing repeatedly
lib.borrow_book("etuu")
lib.borrow_book("etuu")
lib.borrow_book("etuu")

# returning one copy
lib.return_book("etuu")

# final state
lib.show_all_books()

lib.save_report()
```

1) Why is a class better than only lists and functions?

A class is better because it combines data and functions in one place. This makes the code more organized, reusable, and easier to manage compared to using separate lists and functions.

2) Why separate Book and Library into two classes?

Because they represent different things. A Book is a single item, while a Library manages many books. Keeping them separate makes the code cleaner and easier to understand.

3) What responsibilities does each class have?
Book class: stores book details like title and copies
Library class: manages books (add, borrow, return, search, display, save report)


**Question 2 - Student Performance Management System**

Build a student management system using classes, dictionaries, functions, and file handling.

Tasks

a) Create a class named Student with the attributes name, student_id, and grades. The grades attribute must be a dictionary where the key is the course name and the value is the grade.

b) Add the following methods: add_grade(course, grade), update_grade(course, grade), calculate_average(), get_status(), and __str__().

c) A grade must only be accepted if it is between 0 and 20. If the input is invalid, the program must handle it properly.

d) Create a second class named Classroom that stores multiple students.

e) The Classroom class must have methods to add a student, display all students, find the top student, calculate the classroom average, list all students who passed, and list all students who failed.

f) Create at least 5 students, and for each student add at least 4 course grades.

g) Print a full report that shows each student's information, each student's average, pass/fail status, the best student, and the average of the entire classroom.

h) Save the final summary into student_report.txt.

Required written reflection

- Why is a dictionary a good choice for grades?

- Why is calculate_average() a method of the Student class?

- How does the Classroom class improve code organization?

```python
# a) Create a class named Student with the attributes name, student_id, and grades. The grades attribute must be a dictionary where the key is the course name and the value is the grade.

class Student:
  def __init__(self, name, student_id):
      self.name = name
      self.student_id = student_id
      self.grades = {}
# b) Add the following methods: add_grade(course, grade), update_grade(course, grade), calculate_average(), get_status(), and str().

# c) A grade must only be accepted if it is between 0 and 20. If the input is invalid, the program must handle it properly.

  def add_grade(self,grade,course): #serves the same logic for update_grade
    if 0 <= grade <= 20:
        self.grades[course] = grade
    else:
        print("Invalid grade. Must be between 0 and 20.")

  def calculate_average(self):
    if len(self.grades) == 0:
        return 0

    avg = sum(self.grades.values()) / len(self.grades)
    return avg

  def get_status(self):
    avg = self.calculate_average()

    if avg >= 90:
        return "Excellent"
    elif avg >= 50:
        return "Pass"
    else:
        return "Fail"
  def __str__(self):
    avg = self.calculate_average()
    status = self.get_status()

    return (f"Student: {self.name}\n"
            f"ID: {self.student_id}\n"
            f"Grades: {self.grades}\n"
            f"Average: {avg:.2f}\n"
            f"Status: {status}")


s1 = Student("sdf",2)

s1.grades["math"] = 90
s1.grades["Physics"] = 85

print(s1.name)
print(s1.student_id)
print(s1.grades)
print(s1.calculate_average())

print(s1)

```

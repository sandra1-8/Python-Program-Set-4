```python
# a) Create a class named Book with the attributes title, author, and copies, and implement the methods borrow_book(), return_book(), and str().

class Book:
  def __init__(self,title,author,copies):
    self.title = title
    self.author = author
    self.copies = copies

# c) A user must not be able to borrow a book if there are no copies left. The program must print a clear message in that case.

  def borrow_book(self):
    if self.copies > 0:
      self.copies -= 1
      print(f"you borrowed {self.title}")
    else:
      print(f"The book {self.title} is unavailable")

  def return_book(self):
    self.copies += 1
    print(f"1 copy of {self.title} was returned")



  def __str__(self):
    return f"The book {self.title} whose author is {self.author} has {self.copies} copies"

book1 = Book("xsdcvfghj","sdfgh",32)
book2 = Book("werty","tfyui",62)

print(book1)
print(book2)

book1.borrow_book()
book1.borrow_book()
book1.borrow_book()

print(book1)

book2.borrow_book()
book2.borrow_book()
book2.borrow_book()
book2.borrow_book()
print(book2)

```

```python
class Book:
    def __init__(self, title, copies=1):
        self.title = title
        self.copies = copies

    def __repr__(self):
        return f"{self.title} ({self.copies} copies)"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title):
        for book in self.books:
            if book.title == title:
                book.copies += 1
                print(f'Updated: "{title}" now has {book.copies} copies')
                return

        new_book = Book(title)
        self.books.append(new_book)
        print(f'Added new book: "{title}"')

# d) If a user searches for a book that does not exist, the program must handle this safely and display a meaningful message.

    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                print(f'Found: {book}')
                return book
        print("Book not found")
        return None


    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                if book.copies > 0:
                    book.copies -= 1
                    print(f'You borrowed "{title}"')
                else:
                    print("No copies available")
                return

        print("Book not found")

    def return_book(self, title):
        for book in self.books:
            if book.title == title:
                book.copies += 1
                print(f'You returned "{title}"')
                return

        print("Book not found")

    def show_all_books(self):
        if not self.books:
            print("Library is empty")
            return

        print("Library books:")
        for book in self.books:
            print(f"- {book}")

# g) Save the final library report into a text file called library_report.txt.

    def save_report(self, filename="library_report.txt"):
        with open(filename, "w") as file:
            file.write("LIBRARY REPORT\n")
            file.write("=================\n\n")

            for book in self.books:
                file.write(f'Title: {book.title}, Copies: {book.copies}\n')

            file.write("\nEnd of report\n")


# e) Create at least 6 Book objects and add them to the library.

lib = Library()
lib.add_book("fghh")
lib.add_book("sdfgh")
lib.add_book("etuu")
lib.add_book("harry potter")
lib.add_book("lord of rings")
lib.add_book("python basics")

lib.find_book("sdfgh")
lib.borrow_book("etuu")
lib.return_book("fghh")

# f) Simulate a complete scenario that includes displaying all books, borrowing several books, trying to borrow the same book until it becomes unavailable, returning a book, and displaying the final library state.
# initial state
lib.show_all_books()

# borrowing repeatedly
lib.borrow_book("etuu")
lib.borrow_book("etuu")
lib.borrow_book("etuu")

# returning one copy
lib.return_book("etuu")

# final state
lib.show_all_books()

lib.save_report()
```

**Question 2 - Student Performance Management System**

Build a student management system using classes, dictionaries, functions, and file handling.

Tasks

a) Create a class named Student with the attributes name, student_id, and grades. The grades attribute must be a dictionary where the key is the course name and the value is the grade.

b) Add the following methods: add_grade(course, grade), update_grade(course, grade), calculate_average(), get_status(), and __str__().

c) A grade must only be accepted if it is between 0 and 20. If the input is invalid, the program must handle it properly.

d) Create a second class named Classroom that stores multiple students.

e) The Classroom class must have methods to add a student, display all students, find the top student, calculate the classroom average, list all students who passed, and list all students who failed.

Required written reflection

- Why is a dictionary a good choice for grades?

- Why is calculate_average() a method of the Student class?

```python
# a) Create a class named Student with the attributes name, student_id, and grades. The grades attribute must be a dictionary where the key is the course name and the value is the grade.

class Student:
  def __init__(self, name, student_id):
      self.name = name
      self.student_id = student_id
      self.grades = {}
# b) Add the following methods: add_grade(course, grade), update_grade(course, grade), calculate_average(), get_status(), and str().

# c) A grade must only be accepted if it is between 0 and 20. If the input is invalid, the program must handle it properly.

  def add_grade(self,course,grade): #serves the same logic for update_grade
    if 0 <= grade <= 20:
        self.grades[course] = grade
    else:
        print("Invalid grade. Must be between 0 and 20.")

  def calculate_average(self):
    if len(self.grades) == 0:
        return 0

    avg = sum(self.grades.values()) / len(self.grades)
    return avg

  def get_status(self):
    avg = self.calculate_average()

    if avg >= 90:
        return "Excellent"
    elif avg >= 50:
        return "Pass"
    else:
        return "Fail"
  def __str__(self):
    avg = self.calculate_average()
    status = self.get_status()

    return (f"Student: {self.name}\n"
            f"ID: {self.student_id}\n"
            f"Grades: {self.grades}\n"
            f"Average: {avg:.2f}\n"
            f"Status: {status}")

class Classroom:
    def __init__(self):
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def show_all_students(self):
        for s in self.students:
            print(s)

    def top_student(self):
        return max(self.students, key=lambda s: s.calculate_average())

    def classroom_average(self):
        if not self.students:
            return 0
        return sum(s.calculate_average() for s in self.students) / len(self.students)

    def passed_students(self):
        return [s for s in self.students if s.get_status() == "Pass"]

    def failed_students(self):
        return [s for s in self.students if s.get_status() == "Fail"]

    def full_report(self):
        print("CLASS REPORT\n")

        for s in self.students:
            print(s)

        print("\nTop Student:")
        print(self.top_student())

        print(f"\nClass Average: {self.classroom_average():.2f}")

    def save_report(self, filename="student_report.txt"):
        with open(filename, "w") as f:
            f.write("CLASS REPORT\n\n")

            for s in self.students:
                f.write(str(s) + "\n")

            f.write("\nTop Student:\n")
            f.write(str(self.top_student()) + "\n")

            f.write(f"\nClass Average: {self.classroom_average():.2f}\n")


c = Classroom()

s1 = Student("Alice", "S1")
s2 = Student("Bob", "S2")
s3 = Student("Charlie", "S3")
s4 = Student("Diana", "S4")
s5 = Student("Eve", "S5")

# add grades (at least 4 each)
for s in [s1, s2, s3, s4, s5]:
    s.add_grade("Math", 15)
    s.add_grade("Physics", 12)
    s.add_grade("English", 14)
    s.add_grade("CS", 16)

# add to classroom
for s in [s1, s2, s3, s4, s5]:
    c.add_student(s)

c.full_report()
c.save_report()
```

**Reflection**
1. Why is a dictionary good for grades?

A dictionary is good because it stores data as key-value pairs, allowing each course to be directly linked to its grade. This makes access, updates, and organization efficient.

2. Why is calculate_average() in Student?

Because the average belongs to a single student’s data. Each student calculates their own average based on their own grades.

3. How does Classroom improve organization?

The Classroom class manages multiple students and overall operations, separating system-level logic from individual student data, making the code cleaner and easier to maintain.

**Question 3 - Shopping Cart and Invoice Generator**

Create a shopping cart system that uses OOP and produces an invoice.

Tasks

a) Create a class named Product with the attributes name, price, and quantity, and implement the methods subtotal() and __str__().

b) Create a class named Cart that stores several products.

c) The Cart class must include methods for adding a product, removing a product by name, showing all products, computing the total cost, finding the most expensive product, applying a discount, and generating a final invoice.

d) Use the following discount rules: total > 80 -> 10% discount, total > 150 -> 15% discount, otherwise -> no discount.

e) Add at least 7 products to the cart.

f) The invoice must show product name, unit price, quantity, subtotal, original total, discount amount, and final total.

g) If the user tries to remove a product that does not exist, the program must handle it safely.

h) Save the invoice into invoice.txt.

Required written reflection

- Why is subtotal() placed inside the Product class?

- Why is the discount logic placed inside the Cart class?

- How does this design improve readability and maintainability?

```python
# a) Create a class named Product with the attributes name, price, and quantity, and implement the methods subtotal() and str().

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def subtotal(self):
        return self.price * self.quantity

    def __str__(self):
        return f"{self.name} | {self.price}€ x {self.quantity} = {self.subtotal()}€"

# b) Create a class named Cart that stores several products
# c) The Cart class must include methods for adding a product, removing a product by name, showing all products, computing the total cost, finding the most expensive product, applying a discount, and generating a final invoice.

class Cart:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        for p in self.products:
            if p.name == name:
                self.products.remove(p)
                print(f"{name} removed")
                return
        print("Product not found")

    def show_products(self):
        for p in self.products:
            print(p)

    def total_cost(self):
        return sum(p.subtotal() for p in self.products)

    def most_expensive(self):
        return max(self.products, key=lambda p: p.price)

    def apply_discount(self):
        total = self.total_cost()

        if total > 150:
            return total * 0.15
        elif total > 80:
            return total * 0.10
        return 0

    def generate_invoice(self):
        print("\nINVOICE\n")

        for p in self.products:
            print(f"{p.name} | {p.price}€ | Qty: {p.quantity} | Subtotal: {p.subtotal()}€")

        total = self.total_cost()
        discount = self.apply_discount()
        final = total - discount

        print("\n----------------------")
        print(f"Total: {total:.2f}€")
        print(f"Discount: {discount:.2f}€")
        print(f"Final Total: {final:.2f}€")

    def save_invoice(self, filename="invoice.txt"):
        with open(filename, "w") as f:
            f.write("INVOICE\n\n")

            for p in self.products:
                f.write(f"{p.name} | {p.price}€ | Qty: {p.quantity} | Subtotal: {p.subtotal()}€\n")

            total = self.total_cost()
            discount = self.apply_discount()
            final = total - discount

            f.write("\n----------------------\n")
            f.write(f"Total: {total:.2f}€\n")
            f.write(f"Discount: {discount:.2f}€\n")
            f.write(f"Final Total: {final:.2f}€\n")

# e) Add at least 7 products to the cart.

cart = Cart()

cart.add_product(Product("Laptop", 900, 1))
cart.add_product(Product("Mouse", 25, 2))
cart.add_product(Product("Keyboard", 50, 1))
cart.add_product(Product("Headphones", 80, 1))
cart.add_product(Product("USB Cable", 10, 3))
cart.add_product(Product("Monitor", 200, 1))
cart.add_product(Product("Notebook", 5, 5))

# f) The invoice must show product name, unit price, quantity, subtotal, original total, discount amount, and final total.
# g) If the user tries to remove a product that does not exist, the program must handle it safely.

cart.show_products()

cart.remove_product("Tablet")  # not existing → safe message

cart.generate_invoice()

# h) Save the invoice into invoice.txt.

cart.save_invoice()
```

**Reflection**
1. Why is subtotal() in Product?

Because subtotal depends on a single product’s data (price × quantity), so it belongs to the Product class.

2. Why is discount logic in Cart?

Because discounts depend on the total of all products, so it is a responsibility of the Cart, not individual products.

3. How does this design improve readability and maintainability?

By separating responsibilities:

Product handles item-level logic
Cart handles system-level operations

This makes the code cleaner, easier to understand, and easier to extend.

```python
#a) Create a class named Course with the attributes course_name, capacity, and students. The students attribute can be a list of student names or student IDs.
class Course:
    def __init__(self, course_name, capacity):
        self.course_name = course_name
        self.capacity = capacity
        self.students = []
```

```python
#b) Add these methods: register_student(student_name), remove_student(student_name), is_full(), and __str__().
class Course:
    def __init__(self, course_name, capacity):
        self.course_name = course_name
        self.capacity = capacity
        self.students = []

    def register_student(self, student_name):
        if student_name in self.students: #c) A student must not be registered twice in the same course.
            print(f"{student_name} is already registered in {self.course_name}.")
            return

        if self.is_full(): #d) If the course is already full, the program must reject the registration with a clear message.
            print(f"Cannot register {student_name}. {self.course_name} is full.")
            return

        self.students.append(student_name)
        print(f"{student_name} registered in {self.course_name}.")

    def remove_student(self, student_name):
        if student_name in self.students: #c) A student must not be registered twice in the same course.
            self.students.remove(student_name)
            print(f"{student_name} removed from {self.course_name}.")
        else:
            print(f"{student_name} not found in {self.course_name}.")

    def is_full(self):
        return len(self.students) >= self.capacity

    def __str__(self):
        return f"{self.course_name} ({len(self.students)}/{self.capacity} students)"
#c) A student must not be registered twice in the same course.
#d) If the course is already full, the program must reject the registration with a clear message.

```

```python
#e) Create a class named Department that stores multiple courses.
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = {}
```

```python
#f) The Department class must include methods to add a course, display all courses, search for a course, register a student in a course by course name, remove a student from a course, and show which course has the highest number of registered students.
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = {}

    def add_course(self, course):
        self.courses[course.course_name] = course

    def display_courses(self):
        for course in self.courses.values():
            print(course)

    def search_course(self, course_name):
        return self.courses.get(course_name, None)

    def register_student_in_course(self, course_name, student_name):
        course = self.search_course(course_name)
        if course:
            course.register_student(student_name)
        else:
            print("Course not found.")

    def remove_student_from_course(self, course_name, student_name):
        course = self.search_course(course_name)
        if course:
            course.remove_student(student_name)

    def course_with_most_students(self):
        max_course = None
        max_count = -1

        for course in self.courses.values():
            if len(course.students) > max_count:
                max_count = len(course.students)
                max_course = course

        if max_course:
            print(f"Most students: {max_course.course_name} ({max_count})")
```

```python
#g) Create at least 4 courses with different capacities.
c1 = Course("Math", 3)
c2 = Course("Physics", 2)
c3 = Course("CS", 4)
c4 = Course("Biology", 2)
```

```python
#h) Simulate a realistic scenario with many registrations and removals.
dept = Department("Science")

dept.add_course(c1)
dept.add_course(c2)
dept.add_course(c3)
dept.add_course(c4)

# Registrations
dept.register_student_in_course("Math", "Alice")
dept.register_student_in_course("Math", "Bob")
dept.register_student_in_course("Math", "Charlie")
dept.register_student_in_course("Math", "David")  # full

dept.register_student_in_course("Physics", "Alice")
dept.register_student_in_course("Physics", "Bob")

# Duplicate
dept.register_student_in_course("Physics", "Alice")

# Removals
dept.remove_student_from_course("Math", "Bob")

dept.display_courses()
dept.course_with_most_students()
```

```python
#i) Save the final department status into department_status.txt
def save_department(dept, filename):
    with open(filename, "w") as file:
        file.write(f"Department: {dept.name}\n\n")
        for course in dept.courses.values():
            file.write(str(course) + "\n")
            for student in course.students:
                file.write(f" - {student}\n")
            file.write("\n")

save_department(dept, "department_status.txt")
```

```python
#a) Create a NumPy array containing 24 grade values.

import numpy as np

grades = np.array([
    12, 15, 14, 10,
    9, 11, 13, 16,
    18, 17, 15, 14,
    10, 12, 11, 13,
    19, 18, 17, 16,
    8, 9, 10, 11
])

print(grades)
```

```python
#b) Reshape the array into a 6 x 4 matrix where each row represents one student and each column represents one course.

matrix = grades.reshape(6, 4)
print(matrix)
```

```python
#c) Compute the following: overall mean, overall median, overall standard deviation, sum of all grades, mean of each student, mean of each course, and sum of each student.
print("Overall Mean:", np.mean(matrix))
print("Overall Median:", np.median(matrix))
print("Standard Deviation:", np.std(matrix))
print("Sum of all grades:", np.sum(matrix))

print("Mean per student:", np.mean(matrix, axis=1))
print("Mean per course:", np.mean(matrix, axis=0))
print("Sum per student:", np.sum(matrix, axis=1))
```

```python
#d) Using slicing, extract the first two students, the third course for all students, and a submatrix of your choice.
# First two students
first_two = matrix[:2, :]
print("First two students:\n", first_two)

# Third course (column index 2)
third_course = matrix[:, 2]
print("Third course:\n", third_course)

# Submatrix (students 2–4, courses 2–3)
submatrix = matrix[1:4, 1:3]
print("Submatrix:\n", submatrix)
```

```python

#e) Add 2 bonus points to all grades using NumPy operations. No grade may become greater than 20.
bonus_matrix = matrix + 2

# Ensure max value is 20
bonus_matrix = np.clip(bonus_matrix, 0, 20)

print(bonus_matrix)
```

```python
#f) Identify the student with the highest average and the course with the highest average.
student_means = np.mean(matrix, axis=1)
course_means = np.mean(matrix, axis=0)

best_student = np.argmax(student_means)
best_course = np.argmax(course_means)

print("Best student index:", best_student)
print("Best course index:", best_course)
```

```python
#g) Flatten the matrix using both flatten() and ravel(). Then modify one value in the ravel() result and explain what happened.
flat = matrix.flatten()
ravel_array = matrix.ravel()

print("Flatten:", flat)
print("Ravel:", ravel_array)

# Modify ravel result
ravel_array[0] = 100

print("\nAfter modifying ravel:")
print("Ravel:", ravel_array)
print("Original matrix:\n", matrix)
```

2. The big difference here has to do with memory. flatten() creates a brand new copy of the array. ravel() usually just gives you a view of the original data, so if you change a value in the ravel() version, it actually updates your original matrix too.

3. The axis tells NumPy which way to project whatever math operation is being done. If you set axis=0, it calculates things by column, but axis=1 calculates things by row.

```python
import numpy as np
import matplotlib.pyplot as plt

#Question 6 - Broadcasting and Weighted Scores

#(a) Create a 5 x 3 NumPy array representing the grades of 5 students in 3 subjects.
grades = np.array([[85, 92, 78],
                   [90, 88, 56],
                   [30, 92, 77],
                  [46, 100, 83],
                   [23, 56,90]])

#(b) Create a 1D NumPy array containing the subject weights, for example [1, 1.5, 2].
subject_weights = np.array([2, 1.5, 3])

#(c) Use broadcasting to compute the weighted grades.

weighted_grades = grades * subject_weights
print("weighted_grades:\n",weighted_grades)

#(d) Compute the weighted total for each student, the weighted average for each student, the average of each subject before weighting, and the average of each subject after weighting.
weighted_total = np.sum(weighted_grades, axis=1)
print("weighted_total:\n", weighted_total)


average_weighted_grades = np.mean(weighted_grades, axis=1)
print("average_weighted_grades:\n", average_weighted_grades)


average_grades_before_weighting = np.mean(grades, axis=0)
print("average_grades_before_weighting:\n", average_grades_before_weighting)


average_grades_after_weighting = np.mean(weighted_grades, axis=0)
print("average_grades_after_weighting:\n", average_grades_after_weighting)

#(e) Create a bar chart showing the weighted total of each student.
student_names = ["Student 1", "Student 2", "Student 3", "Student 4", "Student 5"]
plt.figure(figsize=(15, 6))
plt.bar(student_names, weighted_total, color="purple")
plt.xlabel("Student")
plt.ylabel("Weighted Total")
plt.title("Weighted Total of Each Student")
plt.show()

print("\n")

#f) Create a line plot comparing subject averages before and after weighting.
subject = ["subject 1", "subject 2", "subject 3"]
plt.figure(figsize=(15, 6))
plt.plot(subject, average_grades_before_weighting, color="blue", label="Before Weighting")
plt.plot(subject, average_grades_after_weighting, color="red", label="After Weighting")
plt.xlabel("Subject")
plt.ylabel("Average Grades")
plt.legend()
plt.title("Comparison of Average Grades Before and After Weighting")
plt.grid(True)
plt.show()

#(g) Both charts must include a title, axis labels, legend where needed, and grid.
#All rules followed as instructed...










```
